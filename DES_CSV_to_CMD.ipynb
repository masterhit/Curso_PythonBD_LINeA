## Leitura dos Daods do DES a partir do arquivo CSV e plot do CMD.


```python
import matplotlib.pyplot as plt
import pandas as pd
from astropy import units as u
from astropy.coordinates import SkyCoord
from dblinea import DBBase

%reload_ext autoreload
%autoreload 2
```

```python
# le dados do csv utilizando pandas
NGC1261_data = pd.read_csv('NGC1261_DES.csv')
```

```python
NGC1261_data.head()
```

```python
NGC1261_data.count()
```

```python
%%time
fontSize = 6

plt.figure(figsize=[6,4], dpi=100)
plt.suptitle("NGC1261 no DES DR2", fontsize=fontSize)
plt.subplot(1,1,1)
plt.plot(NGC1261_data.ra, NGC1261_data.dec, 'k.', alpha=0.1)
plt.xlabel("R.A. (deg)", fontsize=fontSize)
plt.ylabel("Dec. (deg)", fontsize=fontSize)
plt.xticks(fontsize=fontSize)
plt.yticks(fontsize=fontSize)
plt.gca().set_aspect('equal', adjustable='box')
plt.tight_layout()
```

```python
fontSize = 6

plt.figure(figsize=[6,4], dpi=100)
plt.suptitle("NGC1261 no DES DR2", fontsize=fontSize)
plt.hist2d(NGC1261_data.ra, NGC1261_data.dec, bins=50)
plt.xlabel("R.A. (deg)", fontsize=fontSize)
plt.ylabel("Dec. (deg)", fontsize=fontSize)
plt.colorbar(label="Densidade de Pontos")
plt.xticks(fontsize=fontSize)
plt.yticks(fontsize=fontSize)
plt.tight_layout()
```

```python
plt.figure(figsize=[7,5])
plt.title("CMD - NGC1261 no DES DR2", fontsize=12)
plt.hexbin(NGC1261_data.gmr, NGC1261_data.mag_r, gridsize=500, bins='log') #Essa linha cria um grfico hexbin usando Matplotlib, 
#que  particularmente til para visualizar a densidade de pontos de dados em um espao bidimensional.
plt.xlabel("$g - r$", fontsize=fontSize)
plt.ylabel("$g$", fontsize=fontSize)
plt.colorbar(label="Densidade de Pontos")
plt.xticks(fontsize=fontSize)
plt.yticks(fontsize=fontSize)
plt.xlim(-1.5,2.5)
plt.ylim(24,16)
plt.tight_layout()
```
